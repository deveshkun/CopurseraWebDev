Waterfall Model

-> sequential method for development where output of one phase is input for another phase
-> dewvelopment of next phase start after completion of first phase
-> Phases form in shape of waterfall

V Shape Model

-> Phases form in the shape of V
-> Phases going down the left side of V is called verification
-> phases going right side of the V is validation
-> Bottom of V is coding phase

Angile

-> collaberative development process
-> Cycle formation
-> Iterative approach to development
-> It has feedback stage
-> Unit testing is done
